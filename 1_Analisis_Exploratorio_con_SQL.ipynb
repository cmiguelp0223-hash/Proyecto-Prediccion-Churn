```python
import pandas as pd
import sqlite3

csv_file_path = 'WA_Fn-UseC_-Telco-Customer-Churn.csv'

db_file_path = 'teleconnect.db'

# 1. Leer el archivo CSV con pandas
print("Leyendo el archivo CSV...")
df = pd.read_csv(csv_file_path)

# 2. Crear una conexión a la base de datos SQLite
#    (Si el archivo no existe, se creará automáticamente)
conn = sqlite3.connect(db_file_path)

# 3. Cargar los datos del DataFrame a una tabla en la base de datos
#    La tabla se llamará 'churn_data'
print("Cargando datos en la base de datos...")
df.to_sql('churn_data', conn, if_exists='replace', index=False)

print(f"¡Éxito! La base de datos '{db_file_path}' se ha creado con la tabla 'churn_data'.")

db_file_path = 'teleconnect.db'

# 4. Crear un objeto 'cursor' para ejecutar comandos
cursor = conn.cursor()

# 5. Definir la consulta SQL que queremos ejecutar
query = """
    SELECT
        churn,
        COUNT(*) AS total_clientes
    FROM
        churn_data
    GROUP BY
        churn;
"""

# 6. Ejecutar la consulta
cursor.execute(query)

# 7. Obtener todos los resultados de la consulta
results = cursor.fetchall()

# 8. Imprimir los resultados
print("\nResultados:")

for row in results:
    # row[0] es la primera columna (churn), row[1] es la segunda (total_clientes)
    print(f"Churn: {row[0]}, Total Clientes: {row[1]}")

query = """
    SELECT
        churn,
        COUNT(*) AS total_clientes
    FROM
        churn_data
    GROUP BY
        churn;
"""

# 9. Ejecutar la consulta
cursor.execute(query)

# 10. Obtener todos los resultados de la consulta
results = cursor.fetchall()

# 11. Imprimir los resultados
print("\nResultados:")

for row in results:
    # row[0] es la primera columna (churn), row[1] es la segunda (total_clientes)
    print(f"Churn: {row[0]}, Total Clientes: {row[1]}")
query = """
    SELECT
        internetService,
        churn,
        COUNT(*) AS total_clientes
    FROM
        churn_data
    GROUP BY
        internetService, churn
    ORDER BY
        internetService, churn;
"""

results_df = pd.read_sql_query(query, conn)

# 12. Imprimir los resultados
print("\nResultados:")
print(results_df)

# 13. Cerrar la conexión
conn.close()
```
